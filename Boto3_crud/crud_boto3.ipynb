```python
import boto3
import os
import pandas as pd
from dotenv import load_dotenv
```

```python
load_dotenv()
s3=boto3.resource(
    service_name='s3',
    region_name=os.getenv('AWS_DEFAULT_REGION'),
    aws_access_key_id = os.getenv('AWS_ACCESS_KEY_ID'),
    aws_secret_access_key =os.getenv('AWS_SECRET_ACCESS_KEY'))
```

                                                    Create bucket

```python
bucket =s3.create_bucket(Bucket="sanjay-test-bucket1",CreateBucketConfiguration ={'LocationConstraint':'ap-south-1'})
```

                                            Show the all Buckets in S3

```python
for bucket in s3.buckets.all():
    print(bucket.name)
```

                                        Upload the file from the local

```python
s3.Bucket('sanjay-test-bucket1').upload_file(r'C:\Users\sanju\bridgelab_local files\AWS_boto3\anual.csv','anual.csv')
```

```python
for obj in s3.Bucket('sanjay-test-bucket1').objects.all():
    print(obj)
```

                                           Read the data from the s3 bucket file

```python
obj_get=s3.Bucket('sanjay-test-bucket1').Object('anual.csv').get()
pd.read_csv(obj_get['Body'],index_col=0)
```

                                         Download File from local

```python
s3.Bucket('sanjay-test-bucket1').download_file(Key='anual.csv',Filename='download.csv')
pd.read_csv('download.csv',index_col=0)
```

                               Delete the objects in the s3 bucket

```python
bucket=s3.Bucket('sanjay-test-bucket1')
bucket.objects.all().delete()
```

                                      delete the bucket 

```python
bucket=s3.Bucket('sanjay-test-bucket1')
bucket.delete()
```
